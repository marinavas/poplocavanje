```python
def generisi_pol(izl,p):
    r = []
    for i in izl:
        q =affine_h(p,i)
        x,y = q.exterior.coords[0] 
        if(x<2 and x>-2 and y<2 and y>-2):
            r.append(q)
    return r
```

```python
ax = plt.axes()  
ax.set_xlim(0,2)
ax.set_xticks(range(0,3))
ax.set_ylim(0,2)
ax.set_yticks(range(0,3))
ax.set_aspect(1)


slike = generisi_pol(izomgen_p6, pol)
for i in slike:
    patch = PolygonPatch(i, facecolor='#A901DB', edgecolor='#000000', alpha=0.3, zorder=2)
    ax.add_patch(patch) 

print(len(slike))
plt.show()
```
